```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
import folium
```

```python

# read the four datasets
league_df = pd.read_csv("data/league_data.csv")
team_df = pd.read_csv("data/team_data.csv")
social_df = pd.read_csv("data/social_data.csv")
team_df = pd.read_csv("data/team_data.csv")

# set the league_df SeasonStart,SeasonEnd,PlayoffStart,PlayoffEnd to datetime
league_df['SeasonStart'] = pd.to_datetime(league_df['SeasonStart'])
league_df['SeasonEnd'] = pd.to_datetime(league_df['SeasonEnd'])
league_df['PlayoffStart'] = pd.to_datetime(league_df['PlayoffStart'])
league_df['PlayoffEnd'] = pd.to_datetime(league_df['PlayoffEnd'])
```

```python
# create a chart that shows a bar from season start to season end for each season
chart = alt.Chart(league_df).mark_bar().encode(
    x='SeasonStart:T',
    x2='SeasonEnd:T',
    y='LeagueName:N',
).properties(
    width=600,
    height=300
).interactive()


# add the playoffs
playoff_chart = alt.Chart(league_df).mark_bar().encode(
    x='PlayoffStart:T',
    x2='PlayoffEnd:T',
    y='LeagueName:N',
).properties(
    width=600,
    height=300
).interactive()

# combine the two charts
chart + playoff_chart

# set a title
chart + playoff_chart.properties(
    title='Seasons and Playoffs',
    # xlabel="Date",
    # ylabel="League"
)
```
